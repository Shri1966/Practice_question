```python
import numpy as np
import matplotlib.pyplot as plt
```

(a) Draw 100 samples x = [x1 x2] from a 2-dimensional Gaussian distribution with mean [0,0] and identity covariance matrix.

```python
M  = [0,0]
C = [[1,0],[0,1]]
x1,x2 = np.random.multivariate_normal(M, C, 100,).T
plt.plot(x1,x2,'.')
plt.show()
```
